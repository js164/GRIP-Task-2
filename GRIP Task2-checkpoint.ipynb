```python
#importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading Data
iris_df=pd.read_csv('iris.csv')
iris_df
```

```python
#Data Preprocessing
iris_df.isnull().sum()
```

```python
# Finding the optimum number of clusters for k-means classification

iris = iris_df.iloc[:, [0, 1, 2, 3]].values
from sklearn.cluster import KMeans

Sum_of_squared_distance=[]
K=range(1,11)
optimalK=1

for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(iris)
    Sum_of_squared_distance.append(km.inertia_)
    if k>1:
        ratio=Sum_of_squared_distance[k-1]/Sum_of_squared_distance[k-2]
        if ratio < 0.55:
            optimalK=k
```

```python
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'

plt.plot(K,Sum_of_squared_distance)
plt.title('The elbow method')
plt.xlabel('Number of clusters K')
plt.ylabel('Sum of squared distance') 
plt.show()
```

from the above graph, the optimum clusters is where the elbow occurs

So , Number Of Clusters = 3

```python
# Applying kmeans to the dataset

km=KMeans(n_clusters=3)
predict_km=km.fit_predict(iris)
```

```python
# Visualising the clusters

plt.scatter(iris[predict_km == 0, 0], iris[predict_km == 0, 1],s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(iris[predict_km == 1, 0], iris[predict_km == 1, 1],s = 100, c = 'gray', label = 'Iris-versicolour')
plt.scatter(iris[predict_km == 2, 0], iris[predict_km == 2, 1],s = 100, c = 'cyan', label = 'Iris-virginica')

plt.legend()
```

```python
#Plotting with Centroids

plt.scatter(iris[predict_km == 0, 0], iris[predict_km == 0, 1],s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(iris[predict_km == 1, 0], iris[predict_km == 1, 1],s = 100, c = 'gray', label = 'Iris-versicolour')
plt.scatter(iris[predict_km == 2, 0], iris[predict_km == 2, 1],s = 100, c = 'cyan', label = 'Iris-virginica')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')

plt.legend()
```
